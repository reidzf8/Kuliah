```python
from google.colab import drive
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.metrics import accuracy_score, classification_report, mean_absolute_error, mean_squared_error


df = pd.read_csv("kinerja_toko_ritel.csv", sep=";")
df.head()
```

```python
# Pisahkan fitur (X) dan target (y)
X = df.drop(columns=['Keuntungan (Juta)', 'Kategori Kinerja'])  # Fitur
y_class = df['Kategori Kinerja']  # Target untuk klasifikasi
y_reg = df['Keuntungan (Juta)']  # Target untuk regresi
```

```python
# Split data menjadi train dan test
X_train, X_test, y_train_class, y_test_class = train_test_split(X, y_class, test_size=0.2, random_state=42)
X_train, X_test, y_train_reg, y_test_reg = train_test_split(X, y_reg, test_size=0.2, random_state=42)
```

```python
# Standarisasi data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Model KNN untuk Klasifikasi
knn_class = KNeighborsClassifier(n_neighbors=20)
knn_class.fit(X_train_scaled, y_train_class)
y_pred_class = knn_class.predict(X_test_scaled)
```

```python
# Evaluasi Klasifikasi
print("Akurasi Klasifikasi:", accuracy_score(y_test_class, y_pred_class))
print("Laporan Klasifikasi:\n", classification_report(y_test_class, y_pred_class))
```

```python
# Model KNN untuk Regresi
knn_reg = KNeighborsRegressor(n_neighbors=5)
knn_reg.fit(X_train_scaled, y_train_reg)
y_pred_reg = knn_reg.predict(X_test_scaled)
```

```python
# Evaluasi Regresi
mae = mean_absolute_error(y_test_reg, y_pred_reg)
mse = mean_squared_error(y_test_reg, y_pred_reg)
rmse = np.sqrt(mse)

print("Evaluasi Regresi:")
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
```
