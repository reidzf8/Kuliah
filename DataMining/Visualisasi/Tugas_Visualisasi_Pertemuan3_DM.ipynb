```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style seaborn
sns.set(style="whitegrid")
```

```python
# manggillah datnya
data = pd.read_csv('dataset-bersih-elektronik.csv')
data.head()
```

```python
np.sum(data.isnull())
```

```python
# 1. Histogram Distribusi Harga Produk
plt.figure(figsize=(8, 5))
sns.histplot(data["harga"], bins=30, kde=True, color="blue")
plt.title("Distribusi Harga Produk")
plt.xlabel("Harga")
plt.ylabel("Frekuensi")
plt.show()
```

- Sebagian besar produk memiliki harga dalam rentang rendah hingga menengah (di bawah Harga 500.000).
- Terdapat sedikit produk dengan harga tinggi, menunjukkan bahwa kategori produk premium lebih sedikit dibandingkan produk murah atau menengah.
- Distribusi sedikit right-skewed, artinya ada beberapa produk dengan harga sangat tinggi yang menjadi outlier.

```python
# 2. Boxplot Perbandingan Harga per Kategori Produk
plt.figure(figsize=(10, 6))
sns.boxplot(x="kategori", y="harga", data=data, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Perbandingan Harga per Kategori Produk")
plt.xlabel("Kategori Produk")
plt.ylabel("Harga")
plt.show()
```

- Tablet memiliki harga yang cenderung lebih tinggi dibanding kategori lain, dengan median harga yang jauh di atas kategori lain.
- Kategori dengan harga paling rendah terlihat pada beberapa produk seperti Headphone, yang menunjukkan bahwa segmen ini lebih kompetitif di harga murah.
- Beberapa kategori memiliki outlier harga yang sangat tinggi, menunjukkan ada produk premium di kategori tersebut.

```python
# 3. Scatter Plot: Hubungan Harga dan Rating
plt.figure(figsize=(8, 5))
sns.scatterplot(x="harga", y="rating", data=data, alpha=0.6, color="green")
plt.title("Hubungan Harga dan Rating Produk")
plt.xlabel("Harga")
plt.ylabel("Rating")
plt.show()
```

- Tidak terlihat pola yang jelas antara harga dan rating, menunjukkan bahwa harga tidak selalu menentukan kepuasan pelanggan.
- Beberapa produk dengan harga murah tetap memiliki rating tinggi, artinya pelanggan tetap puas meskipun harga rendah.

```python
# 4. Bar Chart: Jumlah Produk per Kategori
plt.figure(figsize=(10, 5))
sns.countplot(x="kategori", data=data, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Jumlah Produk per Kategori")
plt.xlabel("Kategori Produk")
plt.ylabel("Jumlah Produk")
plt.show()
```

- Kategori Smartwatch dan Smartphone memiliki jumlah produk paling banyak, Sedangkan kategori dibawah nya juga memiliki perbedaan selisih jumlah produk yang cukup sedikit, ini menunjukkan bahwa pasar untuk kategori ini lebih kompetitif.
- Kategori Tablet lebih sedikit jumlah produknya, mungkin karena segmentasi pasar yang lebih spesifik atau harga yang lebih mahal.
